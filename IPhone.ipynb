```python
plt.figure(figsize=(10,6))
sns.barplot(x='Product Name', y='Sale Price', data=df, palette='viridis')

plt.title('iPhone Sale Prices')
plt.xticks(rotation=90)   # ✅ correct function
plt.ylabel('Sale Price (INR)')

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Star Rating'],bins=10,kde=True)
plt.title('Distribution of Star Ratings')
plt.xlabel('Star Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Number Of Ratings',y='Number Of Reviews',hue='Star Rating',data=df,palette='cool')
plt.title('Number Of Ratings vs Number of Reviews')
plt.xlabel('Number Of Ratings')
plt.ylabel('Number Of Reviews')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x= 'Ram',data=df,palette='Set2')
plt.title('Distribution of Ram in IPhone')
plt.xlabel('Ram')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Sale Price',y='Star Rating',hue='Ram',data=df,palette='coolwarm')
plt.title('Sale Price vs Star Rating by Ram')
plt.xlabel('Sale Price (INR)')
plt.ylabel('Star Rating')
plt.tight_layout()
plt.show()
```
